```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) #array has more advanced maths functions
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#one axes but in two columns
fig,(ax1,ax2)=plt.subplots(ncols=2,nrows=1,figsize=(7,3))
ax1.scatter(x1,y1)
ax1.annotate('This is important',xy=(10,65),xytext=(30,65),arrowprops=dict(facecolor='black'))
ax2.bar(x2,y2)

#fig,(ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3))
#ax[0].scatter(x1,y1)
#ax[1].bar(x2,y2)

```

```python
#specify the figures layout, ax1 and ax2 are on the left occupying the 1st and 2nd lines, ax3 occupies the right and both rows
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],['ax2', 'ax3']],figsize=(7, 4)) 

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],['ax2', 'ax3']], figsize=(7, 4),layout='constrained')
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )

#fig.tight_layout() #Using this method and the constrained layout methods get the same result
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")
```

```python
#Add error information
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))

ax.plot(x, y2, color = "red")

ax.errorbar(x,y2,yerr = y2_sd,fmt = "none", ecolor='blue', elinewidth=4,capsize=6,capthick=4,errorevery=2)  

#fmt controls how the main line is joined with markers (e.g., "o", "-","o-")
#ecolor specifies the colour for the errorbar only
#errorevery=n, n specifies how many errorbars to take out
```

```python
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
url = "https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png"

#This helps with extracting the image as sometimes wikipedia blocks the download
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "image/*,*/*;q=0.8",
    "Referer": "https://en.wikipedia.org/"
}

response = requests.get(url, headers=headers)       #request the image with the header
image_file = BytesIO(response.content)              #download the image
image = Image.open(image_file)                      #open the image
image
```

```python
#adding the image to the plot
x=np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1=np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

#specify the position for the image
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
0.11,       # y coordinate (ON FIGURE, NOT AXES)
0.15,       # image width
0.35])      # image height

#the box shows where the image will be added


ax_image.imshow(image)        #adding the image to the box
ax_image.axis('off')          #removing the outline of the box
```
