```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50) #generate 50 numbers
y = np.random.randint(0,100,50) #from 0 to 100 with 50 numbers
```

```python
#the figure object and the axes object 
fig, ax = plt.subplots(figsize=(5,3)) #here 5 is the width and 3 is the height
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) #no need for the x value for a histogram
ax.hist(y,edgecolor='black')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Year since start')
fig.tight_layout() #automatically adjust the space between the x-axis and the y-axis to avoid overlapping
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.scatter(x, y, marker='*',color='indigo')
fig.show()
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(x,y, marker='*', color='indigo', linestyle='--',linewidth = 2)
fig.show()
```
